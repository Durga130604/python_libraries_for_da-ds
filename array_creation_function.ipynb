```python
# linspace()                                                       

import numpy as np
a=np.linspace(1,10,5) # will get 5 values between 1 to 10 which will be evenly distributed
print(a)
b=np.linspace(1,10)
print(b)
```

```python
# zeros()

a=np.zeros(5) # to get the dersied amount of zeros
print(a)
b=np.zeros([3,4])
print(b)
c=np.zeros((2,2))
print(c)
```

```python
# ones()

a=np.ones(4)
print(a)
b=np.ones([2,3])
print(b)
c=np.ones((2,2,2))
print(c)
```

```python
# eye()

a=np.eye(5) #(unit matrics)
print(a)
```

```python
# empty()

a=np.empty(7)    #random values
print(a)
```

```python
# arange()

a=np.arange(3)  #random value bydefault starting from 0 till the (given value-1)
print(a)
b=np.arange(1,7)   #random value starting from desired value till the (given value-1)
print(b)
```

```python
# even number between 1-10

a=np.arange(2,10,2)
print(a)
```

```python
# logspace()

print("B\n",np.logspace(2.0,3.0,num=5, base=11)) #BASE=11
print("B\n",np.logspace(2.0,3.0,num=5)) #BASE=10
print("B\n",np.logspace(2.0,3.0,num=5, dtype = int)) #BASE=10
```
