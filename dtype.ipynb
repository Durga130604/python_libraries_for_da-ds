```python
#dtype
import numpy as np
arr=np.array([1,2,3,4])
print(arr.dtype)

arr=np.array(['apple','banana','cherry'])
print(arr.dtype)

arr=np.array([1,2,3,4.0])
print(arr.dtype)
print(arr)
```

```python
#creating array with defined data type
arr=np.array([1,2,3,4],dtype='i4')  
print(arr.dtype)
print(arr)

arr=np.array([1,2,3,4],dtype='i2')  
print(arr.dtype)
print(arr)

arr=np.array([1,2,3,4],dtype='i1')  
print(arr.dtype)
print(arr)

arr=np.array([1,2,3,4],dtype='i3')  #error
print(arr.dtype)
print(arr)
```

```python
#converting data type on existing arrays
arr=np.array([1.1,2.1,3.1,4.1])
print(arr)
print(arr.dtype)

new=arr.astype('i')
print(new)
print(new.dtype)
```

```python
arr=np.array([1,0,3])
newarr=arr.astype(bool)
print(newarr)
print(newarr.dtype)
```
