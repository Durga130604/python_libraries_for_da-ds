```python
# 16-MAR
import matplotlib.pyplot as plt
import numpy as np
data = np.random.normal(size=1000) # size = number of values
plt.hist(data,bins=10) # bins = num of rectangular bar taken to represent the chart
plt.show()
```

```python
# PIE CHART
data = ['apple','banana','cherry','mango']
size = [10,20,30,40]
plt.pie(size,labels=data)
plt.show()
```

```python
# PIE CHART
data = ['apple','banana','cherry','mango']
size = [10,20,30,40]
plt.pie(size,labels=data, autopct='%1.1f%%')
plt.show()
```

```python
# BAR CHART
players = ['Virat', 'Rohit', 'Shikhar', 'Hardik']
runs = [51,87,45,67]
plt.bar(players, runs, color = 'pink')
plt.title('Score Card')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.show()
```

```python
# BAR CHART
players = ['Virat', 'Rohit', 'Shikhar', 'Hardik']
runs = [51,87,45,67]
plt.barh(players, runs, color = 'aquamarine')
plt.title('Score Card')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.show()
```

```python
# LINE CHART
x = [1,3]
y = [8,10]
plt.plot(x,y,marker='o')
plt.title('line chart')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# LINE CHART ANOTHER METHOD
x=np.array([1,8])
y=np.array([3,20])
plt.plot(x,y)
plt.show()
```
