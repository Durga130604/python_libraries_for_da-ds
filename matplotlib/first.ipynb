```python
# 16-MAR
import matplotlib.pyplot as plt
import numpy as np
data = np.random.normal(size=1000) # size = number of values
plt.hist(data,bins=10) # bins = num of rectangular bar taken to represent the chart
plt.show()
```

```python
# PIE CHART
data = ['apple','banana','cherry','mango']
size = [10,20,30,40]
plt.pie(size,labels=data)
plt.show()
```

```python
# PIE CHART
data = ['apple','banana','cherry','mango']
size = [10,20,30,40]
plt.pie(size,labels=data, autopct='%1.1f%%')
plt.show()
```

```python
# BAR CHART
players = ['Virat', 'Rohit', 'Shikhar', 'Hardik']
runs = [51,87,45,67]
plt.bar(players, runs, color = 'pink')
plt.title('Score Card')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.show()
```

```python
# BAR CHART
players = ['Virat', 'Rohit', 'Shikhar', 'Hardik']
runs = [51,87,45,67]
plt.barh(players, runs, color = 'aquamarine')
plt.title('Score Card')
plt.xlabel('Players')
plt.ylabel('Runs')
plt.show()
```

```python
# LINE CHART
x = [1,3]
y = [8,10]
plt.plot(x,y,marker='o')
plt.title('line chart')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# LINE CHART ANOTHER METHOD
x=np.array([1,8])
y=np.array([3,20])
plt.plot(x,y)
plt.show()
```

```python

Fig,a=plt.subplots(2,2)
import numpy as np
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

```python
# 18- MAR
import matplotlib.pyplot as plt
import numpy as np
data = [[30,25,50,20],[40,23,51,17],[35,22,45,19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'g', width=0.25)
ax.bar(X + 0.25, data[1], color = 'y', width=0.25)
ax.bar(X + 0.50, data[2], color = 'r', width=0.25)
```
