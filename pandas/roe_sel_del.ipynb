```python
import pandas as pd
d={'one':pd.Series([1,2,3],index=['a','b','c']),'two':pd.Series([1,2,3,4],index=['a','b','c','d'])}
df=pd.DataFrame(d)
print(df.loc['a']['two'])
print(df.iloc[0,1])
```

```python
#ROW ADDITION
df=pd.DataFrame([[1,2],[3,4],[4,6]],index=['a','b','c'])
print(df)
df2=pd.DataFrame([[5,6],[7,8],[4,6]],index=['a','b','c'])
print(df2)
df3=df.append(df2)
print(df3)
```

```python
#CONCAT FUNCTION
df=pd.DataFrame([[1,2],[3,4],[4,6]],columns=['a','b'])
print(df)
df2=pd.DataFrame([[5,6],[7,8],[4,6]],columns=['a','b'])
print(df2)
df=pd.concat([df,df2],ignore_index=True)
print(df)
```
