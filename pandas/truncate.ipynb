```python
#18-FEB
#TRUNCATE FUNC IN SERIES
import pandas as pd
sr=pd.Series([19,34,56,78,67])
print(sr)
s1=sr.truncate(before=3)
print(s1)
s2=sr.truncate(before=2,after=3)
print(s2)
```

```python
#TRUNCATE FUNC IN DATAFRAME
df=pd.DataFrame({"A":[12,4,5,None,1],
                 "B":[7,2,54,3,None],
                 "C":[14,3,None,2,6]})
index_=['Row_1','Row_2','Row_3','Row_4','Row_5']
df.index=index_
print(df)
result=df.truncate(before='Row_3',after='Row_4')
print(result)
```

```python
#ITERATING OVER ROWS AND COLUMNS IN PANDAS DATAFRAME
#DICTIONARY AS LIST
dict={'name':["aparna","pankaj","sudir","geeku"],
      'degree':["MBA","BCA","M.Tech","MBA"],
      'score':[90,40,80,90]}
df=pd.DataFrame(dict)
#iterating over rows using iterrows() function
for i,j in df.iterrows():
    print(i,j)
```

```python
dict={'name':["aparna","pankaj","sudir","geeku"],
      'degree':["MBA","BCA","M.Tech","MBA"],
      'score':[90,40,80,90]}
for f,r in dict.items():
     print(f,r)
for f in dict:
    for r in dict[f]:
        print(r)
```

```python
#PANDAS DATAFRAME.SORT_VALUES() -1
dict={'name':["aparna","pankaj","sudir","geeku"],
      'degree':["MBA","BCA","M.Tech","MBA"],
      'score':[90,40,80,90]}
#sorting data frame by names
df=pd.DataFrame(dict)
df.sort_values("score",axis=0,ascending=True,
                 inplace=True,na_position='first')   #inplace =changes is permanent
print(df)  
```
