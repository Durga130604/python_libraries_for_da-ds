```python
#19-FEB 
#PANDAS RANDOM FUNC
import numpy as np
import pandas as pd
from numpy.random import randn
df=pd.DataFrame(randn(5,4),['a','b','c','d','e'],['x','y','z','w'])
print(df)
```

```python
#SELECTION OF DATAFRAME
print(df<0)
print(df['x'])
```

```python
print(df[df['x']>0]) #true value of x column and row of that true value will be displayed  df[true]
print(df[df>0])     #df[false=nan]
```

```python
print(df[df['x']>0]['y'])
```

```python
#SELECTION OF DATA FRAME USING LOGICAL OPERATOR
print(df[(df['x']>0) & (df['y']>1)]['y'])
print(df[(df['x']>0) | (df['y']>1)]['y'])
```

```python
#RESET INDEX
import numpy as np
import pandas as pd
from numpy.random import randn
df=pd.DataFrame(randn(5,4),['a','b','c','d','e'],['x','y','z','w'])
print(df.reset_index())
```

```python
outer=['a','a','a','b','b','b']
inner=[1,2,3,1,2,3]
h_index=list(zip(outer,inner))
h_index=pd.MultiIndex.from_tuples(h_index)
df1=pd.DataFrame(randn(6,2))
print(df1)
df1=pd.DataFrame(randn(6,2),h_index)
print(df1)
df1=pd.DataFrame(randn(6,2),h_index,['A','B'])
print(df1)
```
