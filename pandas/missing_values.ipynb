CALCULATIONS WITH MISSING VALUES --

```python
#24-FEB
import pandas as pd
import numpy as np
df=pd.DataFrame({'A':[1,2,np.nan],'B':['a',np.nan,'c'],'C':[10,np.nan,np.nan]})
print(df)
```

```python
#isnull()
print(df['A'].isnull())
```

```python
#ISNULL()
#NOTNULL()
df=pd.DataFrame(np.random.randn(5,3),index=['a','b','c','d','e'],columns=['one','two','three'])
print(df)
df=df.reindex(['p','b','c','d','q'])
print(df.isnull())
print(df.notnull())
```

```python
#SUM()
import pandas as pd
import numpy as np
df=pd.DataFrame({'A':[1,2,np.nan],'B':['a',np.nan,'c'],'C':[10,np.nan,np.nan]})
print(df)
print(df['A'].sum())
print(df['B'].sum())
print(df['C'].sum())
```

```python
#COUNT()
print(df['A'].count())
print(df['B'].count())
print(df['C'].count())
print(df.value_counts())   # returns the count of not null row wise
print(df.count())          # counts ths not null column wise
```

```python
# FILL()
df=pd.DataFrame({'A':[1,2,3],'B':['a',np.nan,'c'],'C':[10,np.nan,20]})
df1=df.fillna(value='Cat')
print(df1)
```

```python
#DROPNA()
df=pd.DataFrame({'A':[1,2,3],'B':['a',np.nan,'c'],'C':[10,np.nan,np.nan]})
print(df)
print(df.dropna())
```

```python
#25-FEB
#fillna()
df=pd.DataFrame({'A':[1,2,3],'B':['a',np.nan,'c'],'C':[10,np.nan,np.nan]})
print(df.fillna(3))
```

```python
#filling mean or sum values of the column in place of nan 
print(df['A'].fillna(value=df['A'].mean()))
print(df['A'].fillna(value=df['A'].sum()))
```

```python
#print(df.fillna(value=df['A'].mean()))
print(df.fillna(value=df['C'].mean()))
```

```python
#thresh() with drop()  thresh- counts least not null value in a column to remove 
df=pd.DataFrame({'A':[1,2,np.nan],'B':[4,5,10],'C':[10,np.nan,np.nan]})
print(df,"\n\n")
print(df.dropna(thresh=2))
```

```python
#REPLACE FUNC
df=pd.DataFrame({'one':[10,20,30,40,50,2000], 'two':[1000,0,30,20,40,50]})
print(df,"\n\n")
print(df.replace({1000:10,2000:0}))
```

GROUPBY() 
1. aggregation   - computing a summary statics
2. transformation  - perform some group-specific operation
3. filteration   - discarding the data with some condition

```python
import pandas as pd
ipl_data={'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings','Riders','Royals','Royals','Riders'],
          'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
          'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
          'points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
print(df)
```

```python
#Splitting data into groups
df1=df.groupby('Year').groups
print(df1,"\n\n")
df2=df.groupby('Team').groups
print(df2,"\n\n")
df3=df.groupby('Rank').groups
print(df3)
```

```python
#df1=df.groupby('Team').max()
#print(df1, "\n\n")
df2=df.groupby('Year').max()
print(df2)
```

```python
# 26-FEB
import pandas as pd
ipl_data={'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings','Riders','Royals','Royals','Riders'],
          'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
          'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
          'points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df1=df.groupby('Team').max()['points']
print(df1)
```

```python
df1=df.groupby('Year')['Rank'].mean()
print(df1)
df2=df.groupby('Year')['Rank'].count()
print(df2)
df3=df.groupby('Year')['Rank'].sum()
print(df3)
df4=df.groupby('Year')['Rank'].std()
print(df4)
```

```python
df1=df.groupby('Year')['Rank'].agg(["std","mean","max"])
print(df1)
```

```python
#score=lambda x:(x+10)
grouped=df.groupby('Team')
score=lambda x: (x+10)               #adding 10 to each rank, eg: 1+10 =11 
print(grouped.transform(score))
```

```python
grouped=df.groupby('Team')
score=lambda x: (x+10)               #adding 10 to each rank, eg: 1+10 =11 
print(grouped['points'].transform(score))
```

```python
#team count should be 3  more than 3
print(df.groupby('Team').filter(lambda x: len(x) >= 3))
```
