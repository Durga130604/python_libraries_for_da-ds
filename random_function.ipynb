```python
# 1.random.rand()
import numpy as np
a=np.random.rand(3)
print(a)
b=np.random.rand(2,2)
print(b)
c=np.random.rand(2,3,4)
print(c)
```

```python
# 2.random.randint()
import numpy as np
a=np.random.randint(4)  # any one value between 0 to given value -1
print(a)
b=np.random.randint(1,10,5)   #any 5 value starting from desired value to given value -1
print(b)
```

```python
# want +ve random numbers
a=np.random.randint(1,10,[2,3])
print(a)
b=np.random.randint(1,10,[2,2,2])
print(b)
c=np.random.randint((3,3,3))
print(c)
```
