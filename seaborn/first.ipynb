```python
import seaborn as sns
df=sns.load_dataset('flights')
df
```

```python
x=df.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
x
```

```python
# HEAT MAP - (dark color - small value, light color - large value)
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.6)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.6,annot=True,fmt='d',cmap='winter') # d means digit , cmap=summer - green 
```

```python
# CLUSTER MAP
sns.clustermap(x)
```

```python
sns.clustermap(x,z_score=0,annot=True,metric='correlation')
```
